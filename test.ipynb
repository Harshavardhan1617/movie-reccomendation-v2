```python
import pandas as pd
import requests
import re
import config

```

```python
df = pd.read_csv('test.csv')
df
```

```python



df['genre'] = ''
df['title'] = ''

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    # Extract the IMDb ID from the r_url column
    imdb_id = re.search('tt\d{7}', row['r_url']).group(0)
    
    # Get the response from OMDB API
    url = f'http://www.omdbapi.com/?i={imdb_id}&apikey={config.OMDB_API_KEY}'
    response = requests.get(url)
    
    # Parse the response as JSON
    data = response.json()
    
    # Extract the genre and title from the response
    genre = data['Genre']
    title = data['Title']
    
    # Add the genre and title to the dataframe
    df.at[index, 'genre'] = genre
    df.at[index, 'title'] = title
df

# # Define a function to extract the IMDB ID from the url
# def get_imdb_id(url):
#     pattern = re.compile(r'tt(\d+)')
#     imdb_id = pattern.search(url).group(0)
#     return imdb_id

# # Define a function to get the genre from the OMDB API
# def get_genre(imdb_id):
#     url = f'http://www.omdbapi.com/?i={imdb_id}&apikey={config.OMDB_API_KEY}'
#     response = requests.get(url)
#     data = response.json()
#     genre = data.get('Genre')
#     return genre
```
